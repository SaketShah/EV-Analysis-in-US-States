```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/shahs/OneDrive/Desktop/Data Analytics/EV in US states/Vehicle Data.csv")
df
```

```python
df = df.drop_duplicates()
df = df.dropna()
```

```python
df
```

```python
df.columns = df.columns.str.strip()
```

```python
num_cols = df.columns.drop('State')
```

```python
num_cols
```

```python
df[num_cols] = df[num_cols].replace(',','',regex = True)
```

```python
df[num_cols] = df[num_cols].apply(pd.to_numeric)
```

```python
df.info()
```

```python
share_cols = ['Electric (EV)','Plug-In Hybrid Electric (PHEV)','Hybrid Electric (HEV)','Gasoline']
for cols in share_cols:
    df[cols + '%'] = (df[cols]/df['Total']) * 100
```

```python
df.head()
```

# Trend & Insights on Alternative Fuels

```python
alt_cols = ['Biodiesel','Ethanol/Flex (E85)','Compressed Natural Gas (CNG)','Propane','Hydrogen','Methanol']
```

```python
total_us_vehicles = df['Total'].sum()
```

```python
alt_summ = df[alt_cols].sum().to_frame('TotalVehicles')
alt_summ['share_%_of_all_vehicles'] = (alt_summ['TotalVehicles']/total_us_vehicles)*100
```

# EV share by state

```python
import matplotlib.pyplot as plt

# Sort by EV %
plot_df = df.sort_values('Electric (EV)%', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(plot_df['State'], plot_df['Electric (EV)%'])
plt.xticks(rotation=90)
plt.ylabel('EV Share (%)')
plt.title('EVs as % of Total Registered Vehicles by State')
plt.tight_layout()
plt.show()
```

# Top 10 EV states by EV count

```python
top10_ev_count = df[['State', 'Electric (EV)']] \
                   .sort_values('Electric (EV)', ascending=False) \
                   .head(10)

plt.figure(figsize=(8, 5))
plt.bar(top10_ev_count['State'], top10_ev_count['Electric (EV)'])
plt.xticks(rotation=45)
plt.ylabel('Number of EVs')
plt.title('Top 10 States by EV Count')
plt.tight_layout()
plt.show()
```

# National fuel mix (EV vs PHEV vs HEV vs Gasoline)

```python
nation = df[['Electric (EV)',
             'Plug-In Hybrid Electric (PHEV)',
             'Hybrid Electric (HEV)',
             'Gasoline',
             'Total']].sum()

labels = ['EV', 'PHEV', 'HEV', 'Gasoline']
values = [nation['Electric (EV)'],
          nation['Plug-In Hybrid Electric (PHEV)'],
          nation['Hybrid Electric (HEV)'],
          nation['Gasoline']]

plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Share of Selected Fuel Types in U.S. Vehicle Fleet')
plt.show()
```
